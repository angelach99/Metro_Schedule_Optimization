```python
import pandas as pd
```

```python
index = [i for i in range(1,21)]
demand = pd.read_excel('Demand_Matrix.xlsx', sheet_name='Matrix')[['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']]
demand.index += 1
demand
```

```python
weekly_demand = {'Orange': 915253.0, 'Green': 892371.5, 'Yellow': 91525.5, 'Blue': 388982.5}
```

```python
def demand_mapping(weekly_demand,df):
    ttl_portion = 0
    for i in [df.Monday,df.Tuesday,df.Wednesday,df.Thursday,df.Friday,df.Saturday,df.Sunday]:
        ttl_portion += sum([j for j in i]) 
    demand_portion = weekly_demand / ttl_portion
    Th =[round(i*demand_portion,0) for i in df.Thursday]
    F =[round(i*demand_portion,0) for i in df.Friday]
    Sat = [round(i*demand_portion,0) for i in df.Saturday]
    Sun = [round(i*demand_portion,0) for i in df.Sunday]
    d = pd.DataFrame({
        "Weekday":Th,
                  "Friday":F,
                  "Saturday":Sat,
                  "Sunday":Sun})
    print(d)
    d.index += 1
    return d
```

```python
orange = demand_mapping(weekly_demand['Orange'],demand)
blue = demand_mapping(weekly_demand['Blue'],demand)
green = demand_mapping(weekly_demand['Green'],demand)
yellow = demand_mapping(weekly_demand['Yellow'],demand)
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('Demand_estimation_Index.xlsx', engine='xlsxwriter')

#Write each dataframe to a different worksheet.

orange.to_excel(writer, sheet_name='orange')
yellow.to_excel(writer, sheet_name='yellow')
blue.to_excel(writer, sheet_name='blue')
green.to_excel(writer, sheet_name='green')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
